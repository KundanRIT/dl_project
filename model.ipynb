```python
import tensorflow as tf
import tensorflow.keras.backend as K
from tensorflow import keras
from tensorflow.keras import Model
from tensorflow.keras.layers import Conv2D, ZeroPadding2D, BatchNormalization, Cropping2D, Activation, Subtract

import datetime
import imageio
import glob
import numpy as np
import datetime
import matplotlib.pyplot as plt
```

```python
# https://stackoverflow.com/questions/41707621/keras-mean-squared-error-loss-layer
def mean_squared_error(y_true, y_pred):
    return K.mean(K.square(y_pred - y_true), axis=-1)
```

```python
# https://stackoverflow.com/questions/44047753/reconstructing-an-image-after-using-extract-image-patches
# Evaluation

patch_size = 42

loaded = keras.models.load_model('model_20191129181140', compile=False)
loaded.compile(optimizer=keras.optimizers.Adam(), loss=mean_squared_error, metrics=['accuracy'])

image = np.asarray(imageio.imread('data/noisy50/0029.png'), dtype='float32') # [x, y, 3]
image = image/255
fig = plt.figure(figsize=(8,8))
plt.imshow(image) # display sample

image = tf.expand_dims(image, 0) # [x, y, 3]
patches = tf.image.extract_patches(images=image, 
                                       sizes=[1, patch_size, patch_size, 1],
                                       strides=[1, patch_size, patch_size, 1],
                                       rates=[1,1,1,1],
                                       padding='VALID') # extract patches
vertical_patch_count = patches.shape[1]
horizontal_patch_count = patches.shape[2]
patches = tf.reshape(patches, [-1, patch_size, patch_size, 3])
denoised_patches = []
for patch in patches:
    denoised_patch = loaded.predict(tf.expand_dims(patch, 0))
    denoised_patches.append(denoised_patch)
denoised_patches = np.asarray(denoised_patches)
denoised_patches = tf.reshape(denoised_patches, [1, vertical_patch_count, horizontal_patch_count, patch_size * patch_size, 3])
denoised_patches = tf.split(denoised_patches, patch_size * patch_size, 3)
denoised_patches = tf.stack(denoised_patches, axis=0)
denoised_patches = tf.reshape(denoised_patches, [patch_size*patch_size, vertical_patch_count, horizontal_patch_count, 3])
denoised_image = tf.batch_to_space(denoised_patches, [patch_size, patch_size], [[0,0],[0,0]])
fig = plt.figure(figsize=(8,8))
plt.imshow(denoised_image[0])

print("psnr", tf.image.psnr(image[0, :denoised_image[0].shape[0], :denoised_image[0].shape[1], :], denoised_image[0], 1))
```

```python
class DeepConv(Model):
    def __init__(self):
        super().__init__()
        self.model_layers = []
        self.model_layers.append(ZeroPadding2D(padding=15, data_format='channels_last'))
        self.model_layers.append(Conv2D(filters=64, kernel_size=(3,3), strides=(1,1),kernel_initializer='Orthogonal', padding='same', activation='relu'))

        self.model_layers.append(Conv2D(filters=64, kernel_size=(3,3), strides=(1,1),kernel_initializer='Orthogonal', padding='same',use_bias = False))
        self.model_layers.append(BatchNormalization())
        self.model_layers.append(Activation('relu'))
        
        self.model_layers.append(Conv2D(filters=64, kernel_size=(3,3), strides=(1,1),kernel_initializer='Orthogonal', padding='same',use_bias = False))
        self.model_layers.append(BatchNormalization())
        self.model_layers.append(Activation('relu'))
        
        self.model_layers.append(Conv2D(filters=64, kernel_size=(3,3), strides=(1,1),kernel_initializer='Orthogonal', padding='same',use_bias = False))
        self.model_layers.append(BatchNormalization())
        self.model_layers.append(Activation('relu'))
        
        self.model_layers.append(Conv2D(filters=64, kernel_size=(3,3), strides=(1,1),kernel_initializer='Orthogonal', padding='same',use_bias = False))
        self.model_layers.append(BatchNormalization())
        self.model_layers.append(Activation('relu'))
        
        self.model_layers.append(Conv2D(filters=64, kernel_size=(3,3), strides=(1,1),kernel_initializer='Orthogonal', padding='same',use_bias = False))
        self.model_layers.append(BatchNormalization())
        self.model_layers.append(Activation('relu'))
        
        self.model_layers.append(Conv2D(filters=64, kernel_size=(3,3), strides=(1,1),kernel_initializer='Orthogonal', padding='same',use_bias = False))
        self.model_layers.append(BatchNormalization())
        self.model_layers.append(Activation('relu'))
        
        self.model_layers.append(Conv2D(filters=64, kernel_size=(3,3), strides=(1,1),kernel_initializer='Orthogonal', padding='same',use_bias = False))
        self.model_layers.append(BatchNormalization())
        self.model_layers.append(Activation('relu'))
        
        self.model_layers.append(Conv2D(filters=64, kernel_size=(3,3), strides=(1,1),kernel_initializer='Orthogonal', padding='same',use_bias = False))
        self.model_layers.append(BatchNormalization())
        self.model_layers.append(Activation('relu'))
        
        self.model_layers.append(Conv2D(filters=64, kernel_size=(3,3), strides=(1,1),kernel_initializer='Orthogonal', padding='same',use_bias = False))
        self.model_layers.append(BatchNormalization())
        self.model_layers.append(Activation('relu'))
        
        self.model_layers.append(Conv2D(filters=64, kernel_size=(3,3), strides=(1,1),kernel_initializer='Orthogonal', padding='same',use_bias = False))
        self.model_layers.append(BatchNormalization())
        self.model_layers.append(Activation('relu'))
        
        self.model_layers.append(Conv2D(filters=64, kernel_size=(3,3), strides=(1,1),kernel_initializer='Orthogonal', padding='same',use_bias = False))
        self.model_layers.append(BatchNormalization())
        self.model_layers.append(Activation('relu'))
        
        self.model_layers.append(Conv2D(filters=64, kernel_size=(3,3), strides=(1,1),kernel_initializer='Orthogonal', padding='same',use_bias = False))
        self.model_layers.append(BatchNormalization())
        self.model_layers.append(Activation('relu'))
        
        self.model_layers.append(Conv2D(filters=64, kernel_size=(3,3), strides=(1,1),kernel_initializer='Orthogonal', padding='same',use_bias = False))
        self.model_layers.append(BatchNormalization())
        self.model_layers.append(Activation('relu'))
        
        self.model_layers.append(Conv2D(filters=64, kernel_size=(3,3), strides=(1,1),kernel_initializer='Orthogonal', padding='same',use_bias = False))
        self.model_layers.append(BatchNormalization())
        self.model_layers.append(Activation('relu'))

        self.model_layers.append(Conv2D(filters=3, kernel_size=(3,3), strides=(1,1), kernel_initializer='Orthogonal',padding='same',use_bias = False))
        self.model_layers.append(Cropping2D(cropping=15, data_format='channels_last'))
        self.subtract = (Subtract())
    
    def call(self, input, **kwargs):
        x = input
        for layer in self.model_layers:
            x = layer(x)
        return self.subtract([input, x])
```

```python
model = DeepConv()
model.compile(optimizer=keras.optimizers.Adam(), loss=mean_squared_error, metrics=['accuracy'])
```

```python
def extract_patches(image):
    image = tf.expand_dims(image, 0)
    patches = tf.image.extract_patches(images=image, 
                                       sizes=[1, patch_size, patch_size, 1],
                                       strides=[1, patch_size, patch_size, 1],
                                       rates=[1,1,1,1],
                                       padding='VALID') # extract patches
    patches = tf.reshape(patches, [-1, patch_size, patch_size, 3])
    return patches
```

```python
noisy_dataset = []
true_dataset = []

for image_path in glob.glob("data/original/*.png"):
    original_image = np.asarray(imageio.imread(image_path), dtype='float32')
    original_image /= 255
    true_dataset.append(extract_patches(original_image))

for image_path in glob.glob("data/noisy50/*.png"):
    noisy_image = np.asarray(imageio.imread(image_path), dtype='float32')
    noisy_image /= 255
    noisy_dataset.append(extract_patches(noisy_image))
    
noisy_dataset = np.asarray(noisy_dataset).reshape(-1, patch_size, patch_size, 3)
true_dataset = np.asarray(true_dataset).reshape(-1, patch_size, patch_size, 3)
dataset = tf.data.Dataset.from_tensor_slices((noisy_dataset, true_dataset))
```

```python
unbatched_dataset = dataset.unbatch()
batched_dataset = unbatched_dataset.batch(batch_size=10)
shuffled_data = batched_dataset.shuffle(buffer_size=80)
test = shuffled_data.take(200).repeat()
train = shuffled_data.skip(200).repeat()
```

```python
now = datetime.datetime.now()
tensorboard_callback = keras.callbacks.TensorBoard(log_dir='logs/log_'.format(now.strftime("%Y%m%d%H%M%S")), histogram_freq=1)
```

```python
model.fit(x=train, validation_data=test, validation_steps=10, steps_per_epoch=2000, epochs=5, callbacks=[tensorboard_callback])
```
